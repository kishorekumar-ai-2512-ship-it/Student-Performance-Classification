```python
#Importing the libraries

from sklearn.preprocessing import LabelEncoder


import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Loading the data
print("Loading the data.......")
data=pd.read_csv(r'C:\Users\kisho\OneDrive\Desktop\AI Mini Projects\Student Performance Classification\Dataset\student_performance_prediction.csv')
data.info()
print("data is loaded.")
```

```python
#Removing the Student ID
data = data.drop(data.columns[0], axis=1)
```

```python
data.head(10)
```

```python
data.isnull().sum()
```

```python
data[['Study Hours per Week','Attendance Rate','Previous Grades']]=data[['Study Hours per Week','Attendance Rate','Previous Grades']].fillna(data[['Study Hours per Week','Attendance Rate','Previous Grades']].mean())
```

```python
data = data.dropna(
    subset=[
        'Parent Education Level',
        'Participation in Extracurricular Activities',
        'Passed'
    ]
)
```

```python
data.isnull().sum()
```

```python
#Encoding the columns

le=LabelEncoder()

data['Participation in Extracurricular Activities']=le.fit_transform(data['Participation in Extracurricular Activities'])
data['Passed']=le.fit_transform(data['Passed'])
```

```python
data = pd.get_dummies(
    data,
    columns=['Parent Education Level'],
    drop_first=True
)
```
